```python
def fun2(a,b,m):
  df1=pd.DataFrame()
  for x in range(11):
    roe=fun(a,b,m,x)

    df1=df1.append(roe,ignore_index=True)
  return df1.fillna(0)
```

```python
a=1597
b=0
m=244944
random.seed(0)
list=[]
df=pd.DataFrame()
for i in range(5):
  list.append(random.randint(1,100))
  s=[]
  temp=(list[i])
  count=0
  freq=[0]*20
  x=[]
  colum=[]
  #table=[]
  for w in range(20):
    x.append(str(w))
    colum.append("["+str(round(w*0.05,2))+","+str(round(w*0.05+0.05,2))+")")
  while temp not in s and count<=10000:
    s.append(temp/m)
    temp=((a*temp+b)%m)
    freq[math.floor(s[count]/0.05)]+=1
    count=count+1
  plt.figure(figsize=(14,5))
  new_row=pd.DataFrame([freq])
  df=df.append(new_row,ignore_index=True)
  plt.bar(x,freq)
  plt.show()
  plt.clf()
df.index=list
df.columns=colum

```
