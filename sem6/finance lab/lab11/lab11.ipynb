```python
def vasicek1(beta,myu,sigma,r0):
    y=np.array([])
    ta=[0]
    Ta=np.arange(1,11,1)
    R0=np.arange(0.1,0.6,0.05)
    for t in ta:
        for T in Ta:
            r=(np.random.normal(r0*np.power(np.e,-beta*t)+myu*(1-np.power(np.e,-beta*t)),np.power((sigma**2)*(1-np.power(np.e,-2*beta*t))/(2*beta),0.5)))
            B=(1-np.power(np.e,-beta*(T-t)))/beta
            A=(myu-((sigma/beta)**2)/2)*(B-(T-t))-(sigma**2/(4*beta))*(B**2)
            p=np.power(np.e,A-B*r);
            y=np.append(y,-np.log(p)/(T-t))
        plt.xlabel('T')
        plt.ylabel(r'y(0,T)')
        plt.plot(Ta,y)  
    plt.show()
    
    
    Ta=np.arange(1,500,1)
    t=0
    for r0 in R0:
        y=np.array([])
        for T in Ta:
            r=(np.random.normal(r0*np.power(np.e,-beta*t)+myu*(1-np.power(np.e,-beta*t)),np.power((sigma**2)*(1-np.power(np.e,-2*beta*t))/(2*beta),0.5)))
            B=(1-np.power(np.e,-beta*(T-t)))/beta
            A=(myu-((sigma/beta)**2)/2)*(B-(T-t))-(sigma**2/(4*beta))*(B**2)
            p=np.power(np.e,A-B*r);
            y=np.append(y,-np.log(p)/(T-t))
        plt.xlabel('T')
        plt.ylabel(r'y(0,T)')
        plt.plot(Ta,y)  
        
    plt.show()
```

```python
def cir1(beta,myu,sigma,r0):
    y=np.array([])
    ta=[0]
    Ta=np.arange(1,11,1)
    R0=np.arange(0.1,0.1,1)
    for t in ta:
        for T in Ta:
            r=r0
            h=(beta**2+2*sigma**2)**0.5
            B=2*(np.power(np.e,(h*(T-t)))-1)/(2*h+(beta+h)*(np.power(np.e,(h*(T-t)))-1))
            A=np.power((2*h*(np.power(np.e,((beta+h)*(T-t)/2)))/(2*h+(beta+h)*(np.power(np.e,(h*(T-t)))-1))),(2*myu*beta/(sigma**2)))
            p=A*np.power(np.e,-B*r);
            y=np.append(y,-np.log(p)/(T-t))
        plt.xlabel('T')
        plt.ylabel(r'y(0,T)')
        plt.plot(Ta,y)  
    plt.show()
    
    
               
    
    
```

```python
def cir2(beta,myu,sigma):
    Ta=np.arange(1,600,1)
    R0=np.arange(0.1,1,0.1)
    t=0
    
    for r0 in R0:
        y=np.array([])
        for T in Ta:
            r=r0
            h=(beta**2+2*(sigma**2))**0.5
            B=2*(np.power(np.e,(h*(T-t)))-1)/(2*h+(beta+h)*(np.power(np.e,(h*(T-t)))-1))
            A=np.power(2*h*(np.power(np.e,((beta+h)*(T-t)/2)))/(2*h+(beta+h)*(np.power(np.e,(h*(T-t)))-1)),(2*myu*beta/(sigma**2)))
            p=A*np.power(np.e,-B*r)
            y=np.append(y,-np.log(p)/(T-t))
        plt.xlabel('T')
        plt.ylabel(r'y(0,T)')
        plt.plot(Ta,y)  
    plt.show()
    
```
