```python
def Asian(myu,sigma,r,T,S0,K,n):
    colours=np.arange(0,10,1)
    calld={}
    putd={}
    for k in K:
        calld[k]=np.array([])
        putd[k]=np.array([])
    fig1,ax1=plt.subplots()
    fig2,ax2=plt.subplots()
    ax1.set_xlabel('t')
    ax1.set_ylabel('Stock prices(risk free)')
    ax2.set_xlabel('t')
    ax2.set_ylabel('Stock prices(real)')
    for j in range(10):
        S=np.array([])
        Sreal=np.array([])
        call=np.array([])
        put=np.array([])
        for i in range(n+1):
            t=(T*i/n)
            w=np.random.normal(0,t**0.5)
            S=np.append(S,(S0*(np.power(np.e,sigma*w+(r-sigma**2/2)*t))))
            w=np.random.normal(0,t**0.5)
            Sreal=np.append(Sreal,(S0*(np.power(np.e,sigma*w+(myu-sigma**2/2)*t))))
        for k in K:
            calld[k]=np.append(calld[k],max(np.mean(S)-k,0))
            putd[k]=np.append(putd[k],max(k-np.mean(S),0))
            ax1.plot(np.arange(0,T+T/n,T/n),S)
            ax2.plot(np.arange(0,T+T/n,T/n),Sreal)
            
    plt.show()
    for k in K:
        print(f'for K={k}:')
        print(f'call values={calld[k]}')
        print(f'put values={putd[k]}')
            
```

```python
def sAsian(sigma,r,T,S0,K,n):
    S=np.array([])
    Sreal=np.array([])
    for i in range(n+1):
        t=(T*i/n)
        w=np.random.normal(0,t**0.5)
        S=np.append(S,(S0*(np.power(np.e,sigma*w+(r-sigma**2/2)*t))))
    call=max(np.mean(S)-K,0)
    put=max(-np.mean(S)+K,0)   
    
    return call,put
```

```python
R=np.arange(0,1,0.05)
Sigma=np.arange(0,1,0.05)
Time=[0,5,0.1]
n=100
S0=100
sigma=0.2
r=0.05
T=0.5
K=112


calls=[]
puts=[]
for r in R:
    call,put=sAsian(sigma,r,T,S0,K,n)
    calls.append(call)
    puts.append(put)
plt.xlabel('r')
plt.ylabel('Option Prices')
plt.plot(R,calls,color='blue')
plt.plot(R,puts,color='red')
plt.show()

calls=[]
puts=[]
K=130
for sigma in Sigma:
    call,put=sAsian(sigma,r,T,S0,K,n)
    calls.append(call)
    puts.append(put)
plt.xlabel('sigma')
plt.ylabel('Option Prices')
plt.plot(R,calls,color='blue')
plt.plot(R,puts,color='red')
plt.show()

calls=[]
puts=[]
K=130
for sigma in Sigma:
    call,put=sAsian(sigma,r,T,S0,K,n)
    calls.append(call)
    puts.append(put)
plt.xlabel('T')
plt.ylabel('Option Prices')
plt.plot(R,calls,color='blue')
plt.plot(R,puts,color='red')
plt.show()

```
