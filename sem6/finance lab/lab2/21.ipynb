```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def call1(M,sigma,r,S0,K,T):
    delt=T/M
    temp1=sigma*np.sqrt(delt)
    temp2=(r-((sigma*sigma)/2))*(delt)
    u=np.power(np.e,temp1+temp2)
    d=np.power(np.e,-temp1+temp2)
    R=np.power(np.e,r*delt)
    p=(R-d)/(u-d)
    if(p<0 or p>1):
        return -1
    call=0
    n=M
    for k in range(n+1):
        call+=(1/np.power(R,n))*(math.comb(n,k)*np.power(p,k)*(np.power(1-p,n-k)*(max(S0*np.power(d,n-k)*(np.power(u,k))-K,0))))
    return call
```
