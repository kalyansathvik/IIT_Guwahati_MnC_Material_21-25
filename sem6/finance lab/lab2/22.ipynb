```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def call3(M,sigma,r,S0,K,T):
    delt=T/M
    temp1=sigma*np.sqrt(delt)
    temp2=(r-((sigma*sigma)/2))*(delt)
    u=np.power(np.e,temp1+temp2)
    d=np.power(np.e,-temp1+temp2)
    R=np.power(np.e,r*delt)
    p=(R-d)/(u-d)
    if(p<0 or p>1):
        return -1
    call=0
    n=M
    for k in range(n+1):
        call+=(1/np.power(R,n))*(math.comb(n,k)*np.power(p,k)*(np.power(1-p,n-k)*(max((S0*np.power(d,n-k)*(np.power(u,k)))**2-K*K,0))))
    return call
```

```python
call3=np.vectorize(call3)
```

```python

def fun3(K):
  S0=np.arange(0,300,5)
  plt.scatter(S0,call3(100,0.3,0.08,S0,100,1),color='red')
  plt.show()

  r=np.arange(8/100,90/100,1/100)
  plt.scatter(r,call3(100,0.3,r,100,100,1),color='red')
  plt.show()

  sigma=np.arange(5/100,80/100,1/100)
  plt.scatter(sigma,call3(100,sigma,0.08,100,100,1),color='red')
  plt.show()
```
