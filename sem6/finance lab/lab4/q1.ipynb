```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def minvarwts(myuv,m,C):
  m=np.array(m)
  n=len(m)
  u=np.array(np.ones(n))
  uT=u.reshape(-1,1)
  C_inv=np.linalg.inv(C)
  mT=m.reshape(-1,1)
  a=u@C_inv@(uT)
  b=u@C_inv@(mT)
  c=m@C_inv@(uT)
  d=m@C_inv@(mT)

  p=d-myuv*b
  q=a*myuv-c
  r=a*d-b*c
  w=(p*(u@C_inv)+q*(M@C_inv))/r

  return np.array(w)

```

```python
M=[0.1,0.2,0.15]
C=[[0.005,-0.010,0.004],[-0.010,0.040,-0.002],[0.004 ,-0.002,0.023]]
```

```python
#1a
```

```python
returns=np.arange(0,0.5,0.0001)
risk=[]
weights=[]
min1=100
min2=100
min3=100
min4=100
min5=100
min6=100
returns15min=0
returns15max=0
returns15min=0
returns10max=0
returns10min=0
returns25max=0
returns25min=0

for myu in returns:
  w=minvarwts(myu,M,C)
  weights.append(w)
  sigma=np.power((w)@(C)@(w.T),0.5)

  if(abs(sigma-0.15)<min1 and myu<0.1):
    min1=abs(sigma-0.15)
    returns15min=myu
  if(abs(sigma-0.15)<min1 and myu>0.1):
    min2=abs(sigma-0.15)
    returns15max=myu

  if(abs(sigma-0.1)<min3 and myu<0.1):
    min3=abs(sigma-0.1)
    returns10min=myu
    w1=w
  if(abs(sigma-0.1)<min4 and myu>0.1):
    min4=abs(sigma-0.1)
    returns10max=myu
    w2=w

  if(abs(sigma-0.25)<min5 and myu<0.1):
    min5=abs(sigma-0.25)
    returns25min=myu
    w3=w
  if(abs(sigma-0.25)<min6 and myu>0.1):
    min6=abs(sigma-0.25)
    returns25max=myu
    w4=w


  risk.append(sigma)

plt.xlabel('risk')
plt.ylabel('returns')
threshold_value = 0.1

index=np.argmin(risk)
threshold_value=returns[index]

colors = np.where(returns > threshold_value, 'r', 'b')

plt.scatter(risk, returns, c=colors)
plt.show()
```
